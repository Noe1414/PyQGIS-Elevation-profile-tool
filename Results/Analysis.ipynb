# Data Analysis

### The goal in this notebook is to compare the data we got with the API to the original data

The data is a 10km course recorded with a garmin watch. The course coordinates have been extracted and used to calculate the elevation. The elevation calculated is then compared to the original one.
The *df1* dataset is the calculated one. the *df2* is the original data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```
